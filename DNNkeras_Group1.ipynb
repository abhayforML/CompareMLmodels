```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split

#kera libraries below
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Activation, Dense, Dropout

from keras.metrics import categorical_crossentropy

#reading data from github and creating DataFrame
import requests
import io

#your sample raw url looks like this
#https://raw.githubusercontent.com/your-username/name-of-the-repository/master/name-of-the-file.csv

#url of project data
url = 'https://raw.githubusercontent.com/abhayforML/CompareMLmodels/main/heart-2.csv'
download = requests.get(url).content

#create df of pandas
df = pd.read_csv(io.StringIO(download.decode('utf-8')))
df.isnull().sum()
df.head()



```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder

le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df.Sex

df['ChestPainType'] = le.fit_transform(df['ChestPainType'])
df.ChestPainType

df['RestingECG'] = le.fit_transform(df['RestingECG'])
df.RestingECG

df['ExerciseAngina'] = le.fit_transform(df['ExerciseAngina'])
df.ExerciseAngina

df['ST_Slope'] = le.fit_transform(df['ST_Slope'])
df.ST_Slope

df.head(10)
```

```python
col_norm = ['Age', 'Sex', 'ChestPainType', 'RestingBP', 'Cholesterol', 'FastingBS',
            'RestingECG', 'MaxHR', 'ExerciseAngina', 'Oldpeak', 'ST_Slope']

#Note last column is dropped which is the Target column

df_norm = df[col_norm]
df_norm.head()

#Normalization below using a function
df1_norm = df[col_norm].apply(lambda x:(x-x.min())/(x.max()-x.min()))

df1_norm.head()

X_data = df1_norm
print(X_data.head())

Y_data = df['HeartDisease']
print(Y_data.head())
```

```python
model = Sequential(name='DNN_Group1_model')
#model.add(Dropout(0.2, input_shape=(11,)))
model.add(Dense(units=64, input_dim=11, activation='relu',name='layer1'))
model.add(Dense(units=64, activation='relu',name='layer2'))
model.add(Dense(units=32, activation='relu',name='layer3'))
model.add(Dense(units=1, activation='sigmoid',name='layer4'))

from tensorflow.keras.optimizers import Adam

# Compile model
#model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
model.compile(loss='mse', optimizer='Adam', metrics=['accuracy'])

# Fit the model
model.fit(X_data, Y_data, epochs=500, batch_size=50, verbose=0)
model.summary()

# evaluate the model
scores = model.evaluate(X_data, Y_data)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

#!pip install ann_visualizer

from ann_visualizer.visualize import ann_viz
ann_viz(model, view=True, filename='projectDNN.gv', title='Project Neural Network')
```
